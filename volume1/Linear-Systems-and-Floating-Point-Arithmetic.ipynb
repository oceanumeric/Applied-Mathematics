In this post, we will start with the problem of solving linear 
systems of the form $Ax = b$. This is a fundamental task in 
linear algebra and is at the core of many engineering 
and scientific applications. Again, the post follows the 
book by Darve, E., & Wootters, M. (2021): Numerical Linear Algebra with Julia. 
However, it will use [JAX](https://jax.readthedocs.io/en/latest/notebooks/quickstart.html) to implement essential algorithms with more detailed explanations. 

We will try to understand __float-point arithmetic__ and __error analysis__
as those are fundamental concepts in numerical analysis. With these tools 
under our belts, we will study __LU factorization__ and discuss __LU factorization with pivoting__
and the __Cholesky factorization__. 

## LU factorization 

In this section, we will see one of the simplest ways to efficiently 
solve a system $Ax = b$, called the LU factorization. This method is 
efficient because it transfers $Ax=b$ into __triangular systems__.
